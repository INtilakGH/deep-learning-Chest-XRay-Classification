**Displaying samples of each class**

```python
def display_sample_images(metadata, labels, num_samples=10):
    plt.figure(figsize=(30, 13))
    for i, label in enumerate(labels):
        # Get the first `num_samples` images for the label
        sample_files = metadata[metadata['Label'] == label][['file_path', 'FILE NAME']].head(num_samples)

        for j, (filepath, filename) in enumerate(zip(sample_files['file_path'], sample_files['FILE NAME'])):
            try:
                img = Image.open(filepath)
                plt.subplot(len(labels), num_samples, i * num_samples + j + 1)
                plt.imshow(img, cmap='gray')
                plt.axis('off')
                plt.title(f"{label}\n{filename}", fontsize=8)  # Add label and file name as title
            except FileNotFoundError:
                print(f"File not found: {filepath}")
    plt.tight_layout()
    plt.show()

# Display 10 images from each label
labels = combined_metadata['Label'].unique()
display_sample_images(combined_metadata, labels, num_samples=10)
```

### **3. Data Processing**

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets (80-20 split)
train_metadata, test_metadata = train_test_split(
    combined_metadata,
    test_size=0.2,
    stratify=combined_metadata['Label'],
    random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_size = (256, 256)
batch_size = 32
```

```python
train_on_gpu = True
```

```python
# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalize pixel values
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    train_metadata,
    x_col='file_path',
    y_col='Label',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_dataframe(
    test_metadata,
    x_col='file_path',
    y_col='Label',
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
```

### **4. Train and Evaluate**

**There are two approaches to consider: building a custom model from scratch or utilizing a pre-trained model. I will begin by developing a custom model, tuning its parameters to optimize performance, and then compare the results with those obtained from a pre-trained model, such as MobileNetV2.**

#### **4.1. Custom model from scratch**

```python
# Build a custom CNN
model_custom = Sequential([
    # Convolutional layers
    Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    # Fully connected layers
    Flatten(),
    Dense(128, activation='relu'),

    Dense(len(train_generator.class_indices), activation='softmax')  # softmax because of multiclass
])
```

```python
model_custom.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model_custom.summary()
```

```python
history_custom = model_custom.fit(
    train_generator,
    validation_data=test_generator,
    epochs=10,
    steps_per_epoch=train_generator.n // batch_size,
    validation_steps=test_generator.n // batch_size
)
```

```python
test_loss_custom, test_accuracy_custom = model_custom.evaluate(test_generator)
print(f"Custom CNN Test Loss: {test_loss_custom}")
print(f"Custom CNN Test Accuracy: {test_accuracy_custom}")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
import time
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)
```

**Now I will apply Batch Normalization**

```python
# Build a custom CNN with BatchNormalization()
model_custom_2 = Sequential([
    # Convolutional layers
    Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    # Fully connected layers
    Flatten(),
    Dense(128, activation='relu'),

    Dense(len(train_generator.class_indices), activation='softmax')  # softmax because of multiclass
])
```

```python
model_custom_2.summary()
```

```python
model_custom_2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
start_time_cust_2 = time.time()
history_custom_2 = model_custom_2.fit(
    train_generator,
    validation_data=test_generator,
    epochs=15,
    steps_per_epoch=train_generator.n // batch_size,
    validation_steps=test_generator.n // batch_size,
    callbacks=[early_stopping]
)
end_time_cust_2 = time.time()
```

```python
total_time_cust_2 = end_time_cust_2 - start_time_cust_2
print(f"Total time taken for model_custom_2: {total_time_cust_2} seconds")
```

```python
test_loss_custom_2, test_accuracy_custom_2 = model_custom_2.evaluate(test_generator)
print(f"Custom CNN Test Loss: {test_loss_custom}")
print(f"Custom CNN Test Accuracy: {test_accuracy_custom}")
```

**The evaluation loss is greater that 1, which indicates some potential issues in the model. I believe BatchNormalization() contributed to it. While Batch Normalization is generally beneficial for training stability and faster convergence, it may have incompatibility with small datasets.</br>
On small datasets, the batch statistics might not generalize well, leading to inconsistencies between training and inference behavior. I am planning to remove it to not over-complicate the model as our dataset is relatively small.
Instead, I will experiment with learning rate and dropout rate.**

```python
from sklearn.model_selection import ParameterGrid

param_grid = {
    'learning_rate': [0.001, 0.00005],
    'dropout_rate': [0.2, 0.5]
}

grid = ParameterGrid(param_grid)
print(f"Number of combinations: {len(grid)}")
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
def build_model(learning_rate, dropout_rate):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
        MaxPooling2D((2, 2)),

        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),

        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),

        Flatten(),
        Dense(128, activation='relu'),
        Dropout(dropout_rate),
        Dense(len(train_generator.class_indices), activation='softmax')
    ])

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
results = []

for params in grid:
    print(f"Training with parameters: {params}")

    model_custom_3 = build_model(learning_rate=params['learning_rate'], dropout_rate=params['dropout_rate'])

    start_time = time.time()
    history = model_custom_3.fit(
        train_generator,
        validation_data=test_generator,
        epochs=10,
        steps_per_epoch=train_generator.n // batch_size,
        validation_steps=test_generator.n // batch_size,
        verbose=0
    )
    end_time = time.time()

    test_loss, test_accuracy = model_custom_3.evaluate(test_generator, verbose=0)

    results.append({
        'params': params,
        'test_loss': test_loss,
        'test_accuracy': test_accuracy,
        'training_time': end_time - start_time
    })
```

```python
import pandas as pd
results_df = pd.DataFrame(results)
print("Grid Search Results:")
display(results_df)
```

```python
best_params = results_df.loc[results_df['test_accuracy'].idxmax()]
print("Best Parameters:")
print(best_params)
```

#### **4.2. Utilizing pre-trained model - MobileNetV2**

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
```

```python
# Load the base model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(256, 256, 3))

# Freeze the base model
base_model.trainable = False

model_MN = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dense(len(train_generator.class_indices), activation='softmax')  # Output layer
])

model_MN.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model_MN.summary()
```

```python
start_time_pre_trained = time.time()
history_MN = model_MN.fit(
    train_generator,
    validation_data=test_generator,
    epochs=10,
    steps_per_epoch=train_generator.n // batch_size,
    validation_steps=test_generator.n // batch_size
)
end_time_pre_trained = time.time()
```

```python
total_time_pre_trained = end_time_pre_trained - start_time_pre_trained
```

```python
total_time_pre_trained
```

```python
test_loss, test_accuracy = model_MN.evaluate(test_generator)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
y_pred_probs = model_MN.predict(test_generator)

y_pred = np.argmax(y_pred_probs, axis=1)

y_true = test_generator.classes
```

### **5. Conclusion**

#### Key findings from the results are as follows:

- Custom CNN without Batch Normalization or Dropout:

Achieved a Test Accuracy of 75.1% and a Test Loss of 0.612, demonstrating decent performance for a simple model.
Training time was not measured but I would expect its training time to be comparable to that of the other two Custom CNN models.
-
Custom CNN with Batch Normalization:

Performed poorly, with a Test Accuracy of 49.2% and a Test Loss of 1.481.
The inclusion of Batch Normalization added computational overhead, as reflected by a Training Time of 1688 seconds, but failed to improve performance, potentially due to its incompatibility with the small dataset.

- Custom CNN with Dropout:

Adding Dropout (rate = 0.2) to the model stabilized performance, achieving a Test Accuracy of 74.5% and a Test Loss of 0.628.
However, training time increased slightly to 1702 seconds, reflecting the computational cost of regularization.

- Pre-trained MobileNetV2:

Outperformed all custom models, achieving the highest Test Accuracy of 84.2% and the lowest Test Loss of 0.422.
Despite its complexity, it required a comparable Training Time of 1674 seconds, showcasing the efficiency of transfer learning.


#### The results demonstrate the clear advantage of using a pre-trained model like MobileNetV2 for image classification tasks, particularly when computational efficiency and performance are priorities. The pre-trained model outperformed all custom CNN variants in both accuracy and loss, emphasizing the value of transfer learning in leveraging pre-trained features from large-scale datasets.

#### While custom models provided a useful foundation for experimentation, their performance plateaued despite parameter tuning and regularization techniques. The introduction of Batch Normalization unexpectedly degraded accuracy, indicating the need for careful integration of such techniques.

#### Points for future enhancements:
- pre-trained models should be the prefered approach
- increasing the number of epochs can be beneficial as the model is randomly initialized and hasn't yet learned meaningful features  
- use a learning rate scheduler to dynamically adjust the learning rate during training
- balance data
- use techniques like oversampling, undersampling, or class weighting to address imbalance.
